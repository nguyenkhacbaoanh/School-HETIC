- comparer le `review_overall` à la moyenne de cette même colonne en retranchant à chaque review la moyenne. Vous pouvez le mettre dans une nouvelle colonne `review_diff` :

```python
# Votre code
df['review_diff'] = df['review_overall'] - df['review_overall'].mean()
df.head()
```

- combien de bières de type `Pumpkin Ale` y'a-t-il dans le dataset ? Vous pouvez regarder du côté de la méthode `map` et des fonctions `lambda` de Python.

```python
# Votre code
# sum(df['beer_style'].apply(lambda x: x=='Pumpkin Ale'))
# sum(map(lambda x: x=='Pumpkin Ale',df['beer_style']))
print(pd.value_counts(df["beer_style"])["Pumpkin Ale"])
```

- quel est la `review_overall` moyenne par brasserie (`brewery_name`) ? Regardez du côté de `groupby`.

```python
# Votre code
df.groupby('brewery_name')['review_overall'].mean()
```

- faire la même chose avec `median` en plus en **une seule ligne** :

```python
# Votre code
df.groupby('brewery_name')['review_overall'].agg(['mean', 'median', 'min'])
```
