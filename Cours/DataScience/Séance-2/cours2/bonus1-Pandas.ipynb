Nous allons nous concentrer sur l'import de fichier Excel et le cleaning de données afin de pouvoir mieux les exploiter dans un `DataFrame`.

Pour ce faire, téléchargez le fichier `gap_hist.xlsx` à [cette adresse](https://www.bea.gov/national/xls/gap_hist.xlsx), importez la _table 6_ ("Motor Vehicle Unit Reteail Sales") et nettoyez le `DataFrame`.

```python
# Votre Code
import pandas as pd
def map_(x,y):
    return str(x) + "-" + str(y)

def clean_data(path, select_table):
    with pd.ExcelFile(path) as xlsx:
        df = pd.read_excel(xlsx, select_table, header=3).reset_index()
    df.drop(index=df.loc[::13,:].index, inplace=True)
    df['level_0'] = list(map(map_, df['level_0'], df['level_1'].astype('int16', inplace=True)))
    df.drop(['level_1'], axis=1, inplace=True)
    df.rename(columns={'level_0':'Date'}, inplace=True)
    for name_col in df.columns:
        if "Thousands" in name_col:
            df[name_col] *= 1000
        name = name_col.split("--")[0].strip()
        if 'annual rates' in name_col:
            name = name + '_' + 'seasonally'
        else:
            name = name
        name = "_".join(name.split(" "))
        df.rename(columns={name_col:name}, inplace=True)
    return df

df = clean_data('gap_hist.xlsx','Table 6')
```
