```python
import pandas as pd
import matplotlib
import seaborn as sns
%matplotlib inline
sns.set(rc={'figure.figsize':(16,8)})
```

```python
salary = pd.read_csv("salaries.csv", index_col=False)
salary.head()
```

```python
salary['education'].value_counts().plot.bar()
```

```python
salary['salary'].plot.hist()
```

```python
# scatter plot: nuage des point
sns.scatterplot('experience', 'salary', data=salary, hue='education')
```

```python
sns.boxplot('education', 'salary', data=salary, hue='management')
```
