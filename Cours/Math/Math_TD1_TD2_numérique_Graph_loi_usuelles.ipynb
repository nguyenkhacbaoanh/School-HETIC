# Loi uniform: $\mathscr{U}(a,b)$

- Equation de la densité: $$f_{x\in[a,b]}(x) = \frac{1}{b - a}$$
- Equation de la répartition: $$F_{x\in[a,b]}(x) = \frac{x - a}{b - a}$$

```python
def f_uniforme(x, a, b):
    if (x>=a and x <=b):
        return 1/(b-a)
    else:
        return 0

def F_uniforme(x, a, b):
    if x < a: 
        return 0
    elif (x>=a and x <=b): 
        return (x-a)/(b-a)
    else: 
        return 1
```

# Loi normale $\mathscr{N}(m,\sigma)$
---

- Equation de la densité: $$\frac{1}{\sigma\sqrt{2\pi}}\exp{-\frac{(x - m)^2}{2\sigma^2}}$$
- Equation de la partition: $$F_X(x) = \frac{1}{2}(1+ erf(\frac{x-m}{\sigma\sqrt{2}}))$$

```python
def loi_normale(x , m, sigma):
    return (1./(sigma*np.sqrt(2.0*np.pi)))*(np.exp(-(((x - m)**2)/(2.0*sigma**2))))
```

```python
x = np.linspace(-1,1, 500)
y = loi_normale(x, 0, 1)
y1 = loi_normale(x, 0, 1.5)
y2 = loi_normale(x, 0, 0.2)
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3)
ax1.plot(y)
ax2.plot(y1)
ax3.plot(y2)
```

# Loi exponentielle $\mathscr{E}(\lambda)$
---

- Equation de la densité: $$f_{X \in \mathbb{R}^+}(x) = \lambda e ^{-\lambda x}$$
- Equation de la répartition: $$F_{X \in \mathbb{R}^+}(x) = 1 - e ^{-\lambda x}$$


```python
def loi_exponentielle(x, lambda_):
    if x < 0:
        return 0
    else:
        return lambda_*np.exp(-lambda_*x)

def repartition_exponentielle(x, lambda_):
    if x <0:
        return 0
    else:
        return (1 - (np.exp(-lambda_*x)))
```

```python
x = np.linspace(0,5,5000)
y = [loi_exponentielle(i, 1.5) for i in x]
y1 = [repartition_exponentielle(i, 1.5) for i in x]
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
ax1.plot(y)
ax2.plot(y1)
```
