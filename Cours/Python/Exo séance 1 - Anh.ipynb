```
# Un test de parité d’un entier
def exo():
    valid = False # c'est la variable boolean qui nous permet de valider nos constructions
    while valid == False: # boucler jusqu'à variable valid change la valeur
        nombre = input("Nombre d'entier:" ) # le saisi de l'utilisateur
        try: # mos conséquences risque à erreur
            nombre = '.'.join(nombre.split(',')) # découpant le chain de caractère à point virgure et se réjointant en point, pour la consigne 3
            nombre = float(nombre) # pour la consigne 1 et 2, si chain caractère est possible de faire cast fonction
            valid = True
        except ValueError:
            print("C'est pas le nombre réel")
        finally:
            print(nombre)
exo()
```

```
def random_value(saut=1):
    try:
        valid = False
        while valid == False:
            print("Veuillez saisir le fouchet de nombre que vous voullez")
            debut = input("le nombre commencé: ")
            fin = input("le nombre terminé: ")
            saut = saut # ce parametter nous permet de choisir l'écarte de nombre précédent et suivant, vaut 1 par défaut
            if int(debut)>int(fin):
                print("\nLe nombre terminé doit être supérieur ou égale au nombre commencé\n")
            else: valid = True
        import random
        tableau_value = range(int(debut), int(fin)+1 , saut)
        print("Toute les valeurs possibles dans votre fourchet:")
        for i in tableau_value:
            print("{} \t".format(i), end="")
        value = random.randrange(int(debut), int(fin)+1, saut)
        value_flottant = float(value)
        print("\nLa valeur aléatoire: {:d} \nLa valeur aléatoire en flottant {:2f}".format(value, value_flottant))
    except ValueError as e:
        print("\nVeuillez saisir le nombre entier, svp!\n")
        random_value()
        
random_value()
```
