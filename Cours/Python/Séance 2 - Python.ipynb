```
seq = list([1,2,3])
print(hasattr(seq,"__iter__")) # verifie l'objet est iterable
print(type(seq.__iter__()))
try:
    next(seq) # try to method next de iterateur
except:
    print(f"{seq} est iterable mais n'est pas un iterateur")
iter_seq = iter(seq)
print(hasattr(iter_seq,"__iter__"))
print(next(iter_seq))
print(next(iter_seq))
```

```
dic = dict({
    "Hello": "Bonjour",
    "Say": "Parler"
})
print(hasattr(dic, "__iter__"))
print(hasattr(dic, "__next__"))

iter_dic = iter(dic)
print(hasattr(iter_dic, "__next__"))
print(next(iter_dic))
print(next(iter_dic))
```

```
import random
def gen_random():
    tmp = random.randint(1,12)
    while tmp != 6:
        yield tmp
        tmp = random.randint(1,12)
    yield tmp # recupère le numéro 6 quand le boucle s'épuise

gr = gen_random()
# utilse itérateur et gestion des exceptions
#while True:
#   print(next(gr))

for i in gr:
    print(i)
```
